```python
# ë°°ì—´ì˜ í¬ê¸° N, ìˆ«ìê°€ ë”í•´ì§€ëŠ” íšŸìˆ˜ M, ì—°ì† ê°€ëŠ¥ íšŸìˆ˜ K ì…ë ¥
n, m, k = map(int, input().split())
print(f"n, m, k: {n, m, k}")

# ë°°ì—´ ì…ë ¥
array = list(map(int, input().split()))
print(f"array: {array}")

# ê°€ì¥ í° ìˆ˜ì™€ ë‘ ë²ˆì§¸ë¡œ í° ìˆ˜ ì°¾ê¸°
array.sort(reverse=True)
first = array[0]
second = array[1]

# ê°€ì¥ í° ìˆ˜ì˜ ì—°ì† ê°€ëŠ¥ íšŸìˆ˜ë§Œí¼ ë”í•œ ê°’ì„ ìµœëŒ€ë¡œ ë”í•  ìˆ˜ ìˆëŠ” íšŸìˆ˜
first_cnt = m // k

# ë‘ ë²ˆì§¸ í° ìˆ˜ë¥¼ ë”í•´ì•¼ í•˜ëŠ” íšŸìˆ˜
second_cnt = m % k

result = (first * k * first_cnt) + (second * second_cnt)
print(result)
```

```python
def solution(n, m, k, array):
    # ê°€ì¥ í° ìˆ˜ì™€ ë‘ ë²ˆì§¸ë¡œ í° ìˆ˜ ì°¾ê¸°
    array.sort(reverse=True)
    first = array[0]
    second = array[1]

    # ê°€ì¥ í° ìˆ˜ê°€ ì—°ì†ìœ¼ë¡œ ë”í•´ì§ˆ ìˆ˜ ìˆëŠ” íŒ¨í„´ì˜ ë°˜ë³µ íšŸìˆ˜
    first_pattern_cnt = m // k

    # ë‘ ë²ˆì§¸ í° ìˆ˜ë¥¼ ë”í•´ì•¼ í•˜ëŠ” íšŸìˆ˜ëŠ” ë‚¨ì€ íšŸìˆ˜
    second_cnt = m % k

    return (first * k * first_pattern_cnt) + (second * second_cnt)
```
